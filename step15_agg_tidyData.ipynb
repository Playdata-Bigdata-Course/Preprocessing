### 데이터 집계 & 정돈된 형태로 데이터 재구성하는 연습

```python
import numpy as np
import pandas as pd
```

```python
flights = pd.read_csv("0.dataSet/flights.csv")
flights.head()
```

```python
flights.info()
```

```python
flights.columns
```

```python
# 항공사 별(AIRLINE)로 도착 지연시간(ARR_DELAY)에 대한 평균(mean) 연산 집계
# 딕셔너리 형식 권장 
flights.groupby('AIRLINE').agg({'ARR_DELAY' : 'mean'}).head()
```

```python
flights.groupby('AIRLINE')['ARR_DELAY'].agg('mean').head()
```

```python
flights.columns
```

```python
# 항공사별(AIRLINE) 주중(WEEKDAY) 취소건수(CANCELLED), 우회 운항 건수(DIVERTED)와 비율
# 비율 - 합, 평균
flights.groupby(['AIRLINE' , 'WEEKDAY'])['CANCELLED', 'DIVERTED'].agg(['sum', 'mean'])
```

```python
flights.WEEKDAY
```

```python
flights.columns
```

```python
# 각 출발지(ORG_AIR)와 목적지별(DEST_AIR) 전체 운항 편수(sum), 운항 취소(CANCELLED)된 건수와 
# 비율(sum, mean), 비행 시간(AIR_TIME)의 평균과 분산('var')

group_cols = ['ORG_AIR', 'DEST_AIR']
agg_dict = {'CANCELLED':['sum', 'mean']  ,  'AIR_TIME':['mean', 'var'] }

air_info = flights.groupby(group_cols).agg(agg_dict)
air_info
```

```python
air_info.columns
```

```python
level0 = air_info.columns.get_level_values(0)
level0
```

```python
level1 = air_info.columns.get_level_values(1)
level1
```

```python
air_info.columns = level0 + '_' + level1
air_info.head()
```

```python
air_info.reset_index()
```

```python
# as_index=False : 행 index를 group by 시에 제거하는 설정

group_cols = ['ORG_AIR', 'DEST_AIR']
agg_dict = {'CANCELLED':['sum', 'mean']  ,  'AIR_TIME':['mean', 'var'] }

air_info = flights.groupby(group_cols, as_index=False).agg(agg_dict)
air_info
```

### 정돈된 데이터(Tidy Data) 구성해 보기 

1. 분석이 용이한 구조 의미
2. 정돈된 데이터 특징
>각 변수는 열을 형성<br>
>각 관측값은 행을 형성<br>
>각 관측된 단위별로 별도의 table로 구성

```python
fruit = pd.read_csv('0.dataSet/fruit.csv')
fruit
```

```python
fruit = pd.read_csv('0.dataSet/fruit.csv', index_col='State')
fruit
```

```python
# Apple	Orange	Banana 이름은 컬럼명이 아닌 data롤 변환(정돈)
# stack()사용해서 모든 열의 이름을 수직으로 재구성
fruit.stack()
```

```python
f_tidy = fruit.stack().reset_index()
```

```python
f_tidy
```

```python
f_tidy.columns = ['state', 'fruit', 'weight']
```
